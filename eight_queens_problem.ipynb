```python
import random
```

```python
def is_valid(board):
    print(f"is_valid board: {board}")
    n = len(board)
    print(f"is_valid len: {n}")
    # Check if no two queens threaten each other
    for i in range(n):
        for j in range(i+1, n):
            if board[i] == board[j] or abs(board[i] - board[j]) == abs(i - j):
                return False
    return True
```

```python
def fitness(board):
    n = len(board)
    clashes = 0
    for i in range(n):
        for j in range(i+1, n):
            if board[i] == board[j] or abs(board[i] - board[j]) == abs(i - j):
                clashes += 1
    print(f"fitness clashes: {clashes}")
    print(f"fitness return value: {28 - clashes}")
    return 28 - clashes  # 28 is the maximum number of non-attacking pairs
```

```python
def random_solution(n):
    random_solution =  [random.randint(0, n-1) for _ in range(n)]
    print(f"random_solution: {random_solution}")
    return random_solution
```

```python
def crossover(parent1, parent2):
    print(f"crossover parent1: {parent1}")
    print(f"crossover parent2: {parent2}")
    n = len(parent1)
    crossover_point = random.randint(1, n - 2)
    print(f"crossover crossover_point: {crossover_point}")
    child = parent1[:crossover_point]
    print(f"crossover child: {child}")
    for gene in parent2:
        if gene not in child:
            child.append(gene)
    print(f"crossover return child: {child}")
    return child
```

```python
def mutate(solution, mutation_rate):
    n = len(solution)
    if random.random() < mutation_rate:
        print(f"mutate random.random(): {random.random()}")
        i, j = random.sample(range(n), 2)
        solution[i], solution[j] = solution[j], solution[i]
        print(f"mutate i: {i}")
        print(f"mutate j: {j}")
        print(f"mutate solution[i]: {solution[i]}")
        print(f"mutate solution[j]: {solution[j]}")
    print(f"mutate return solution: {solution}")
    return solution
```

```python
def genetic_algorithm(population_size, mutation_rate, generations):
    n = 8  # Number of queens
    population = [random_solution(n) for _ in range(population_size)]
    print(f"genetic_algorithm population: {population}")
    for generation in range(generations):
        print(f"generation: {generation}")
        population = sorted(population, key=lambda x: -fitness(x))  # Sort in descending order
        print(f"population sorted: {population}")
        if fitness(population[0]) == 28:  # Found a solution
            return population[0]
        new_population = population[:population_size // 2]  # Select top 50%
        print(f"Select top 50%: {new_population}")
        for _ in range(population_size // 2):
            parent1, parent2 = random.choices(new_population, k=2)
            child = crossover(parent1, parent2)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        population = new_population
    return None  # No solution found
```

```python
population_size = 100
mutation_rate = 0.1
generations = 1000
solution = genetic_algorithm(population_size, mutation_rate, generations)
    
if solution:
    print("Found a solution:")
    print(solution)
else:
    print("No solution found.")
```
